```python
import random
import matplotlib.pyplot as plt
from matplotlib import style

def createPlots(xrange, yrange):
    xs = []
    ys = []
    
    for i in range(xrange):
        x = i
        y = random.randrange(yrange)
        
        xs.append(x)
        ys.append(y)
    return xs, ys

fig = plt.figure()
fig.subplots_adjust(hspace=0) 
fig.subplots_adjust(wspace=0) 

gridsize = (1,3)
x,y = createPlots(10,20)
#                           starting point
ax1 = plt.subplot2grid(gridsize, (0,0), rowspan=1)
plt.title('Graph 1')
ax1.plot(x, y, c = 'g', label='y1')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()

ax2 = plt.subplot2grid(gridsize, (0,1), rowspan=1, sharey = ax1)
# ax2.set_title('Graph 2')
x,y = createPlots(10,30)
ax2.plot(x,y, c = 'red', label='y2')
ax2.grid(True)
plt.setp(ax2.get_yticklabels(), visible=False)
ax2.legend()

ax3 = plt.subplot2grid(gridsize, (0,2), rowspan=1, sharey = ax1, sharex = ax2)
# ax3.set_title('Graph 3')
x,y = createPlots(30,10)
ax3.plot(x, y, c = 'blue', label='y3')
ax3.grid(True)
plt.setp(ax3.get_yticklabels(), visible=False)
ax3.legend()

#    plt always applies only to the last ax
plt.xlabel('x')
plt.title('Graphs')
fig.tight_layout()
```
